# Importando bibliotecas

```python
import tabula
```

# Exibindo e tratando as tabelas

## tab1 e tab2

```python
listas_tabelas = tabula.read_pdf("../desafio/dados/censoEducacaoMaranhao.pdf", pages = "79")
for tabela1 in listas_tabelas:
    display(tabela1)
```

### Eliminando valores nulos

```python
A = tabela1.shape[0]
tabela1.dropna(subset = ['Unnamed: 0', 'Unnamed: 1', 'Maranhão', 'Unnamed: 2', 'Nordeste', 'Unnamed: 3','Unnamed: 4', 'Brasil', 'Unnamed: 5'], inplace = True)
B = tabela1.shape[0]
A - B
```

### Removendo colunas e linhas

```python
tabela1 = tabela1.drop(columns=['Nordeste','Unnamed: 2','Unnamed: 3','Unnamed: 4','Brasil','Unnamed: 5'])
tabela1 = tabela1.drop(47, axis=0)
tabela1
```

```python
tabela1[["Urbana","Rural"]] = tabela1["Maranhão"].str.split(" ", expand = True)
tabela1 = tabela1.drop(columns=['Maranhão'])
display(tabela1)
```

### Renomeando colunas

```python
tabela1 = tabela1.rename(columns={'Unnamed: 0' : 'Indicador', 'Unnamed: 1' : 'Total'})
tabela1
```

### Salvando tabelas da pagina 79


####  Indicadores TODAS AS REDES do ensino médio segundo a rede de ensino no Maranhão – 2020

```python
tab1 = tabela1[:29]
tab1
```

```python
tab1.to_csv('dados limpos/dados_todas_as_redes.xls', index=False)
```

####  Indicadores REDE PÚBLICA do ensino médio segundo a rede de ensino no Maranhão – 2020

```python
tab2 = tabela1[29:]
tab2
```

```python
tab2.to_csv('dados limpos/dados_rede_publica.xls', index=False)
```
